### Part 1: Colors


Matplotlib assigns colors by default, but you can choose the colors you want as well. Matplotlib has many ways to represent colors. These three methods are the most common:

1. Single-character representation for some of the primary colors, e.g., "b" for blue. You can see in the table below that many colors do not have the option for single-character representation.  You should check current matplotlib documentation to verify the latest capabilities.

2. The color name, e.g., "blue".

3. The common RGB (red, green and blue) color model used in digital imaging. The colors are represented using the hexadecimal number system, also called base-16 or sometimes just hex. Hexidecimal uses 16 unique symbols to represent values. Those symbols are 0-9 and A-F, e.g., "0000FF " for blue.

Many matplotlib methods allow color modification, usually with a straightforward parameter such as `color`. You add color to line plots like this:

To demonstrate, we'll use each of these representations to color the line for a different company's stock:

### Part 2: Legends

Regardless of the type of chart or visualization you are creating, it is important to include descriptive elements like labels and a legend. The legend is the small gray box on the left side of the graph in the introduction section that shows the color and name of each line on the graph. This plot component helps the viewer to interpret the plot. 


 To incorporate a legend into the plot, we'll use the `legend()` method. There are two parts to adding a legend.
 - First, you need to label each line so that matplotlib knows what to put in the legend. 

 `plt.plot(x_values, y_values, label='name_to_appear_in_legend')` 

- Next, add the `legend()` method without any parameters. Matplotlib infers the plot elements that need to be included in the legend as long as they are labeled.

`plt.legend()`


The following example demonstrates adding a legend to the stocks plot, giving those executives the ability to know which line belongs to which company:

```python
# Rocket line for plot colored blue
plt.plot(stocks['day'],stocks['rocket'],color='b',label='Rocket')
 
# Biotri line for plot colored red
plt.plot(stocks['day'],stocks['biotri'],color='red',label='Biotri')
 
# Wheatpharm line for plot colored magenta
plt.plot(stocks['day'],stocks['wheatpharm'],color='#00FF00',label='Wheatpharm')
 
# plot title
plt.title("Comparison of Wheatpharm, Rocket & Biotri")
 
# label x-axis
plt.xlabel("Days (since IPO)")
 
# label y-axis
plt.ylabel("Price (USD)")
 
# add a legend
plt.legend()
 
# display figure
plt.show()
```

What if you want the legend somewhere else on the plot?! Instead of leaving it up to matplotlib to determine the legend placement on the plot automatically, you can move the location of the legend. Use the loc parameter like this:

`plt.legend(loc='position value')`

The position value choices are:

||||
| -------------- | -------------- | -------------- |
|`'upper left'`  |`'lower left'`  |`'center left'` |
|`'upper right'` |`'lower right'` |`'center right'`|
|`'upper center'`|`'lower center'`| |

Let's move the position of the legend to the center-left from the upper right. We need that space for something else later. 

```python
# Rocket line for plot colored blue
plt.plot(stocks['day'],stocks['rocket'],color='b',label='Rocket')
 
# Biotri line for plot colored red
plt.plot(stocks['day'],stocks['biotri'],color='red',label='Biotri')
 
# Wheatpharm line for plot colored green
plt.plot(stocks['day'],stocks['wheatpharm'],color='#00FF00',label='Wheatpharm')
 
# plot title
plt.title("Comparison of Wheatpharm, Rocket & Biotri")
 
# label x-axis
plt.xlabel("Days (since IPO)")
 
# label y-axis
plt.ylabel("Price (USD)")
 
# add a legend
plt.legend(loc='center left')
 
# display figure
plt.show()
```

### Part 3: Annotations

Annotations are used to call attention to a particular feature of a plot in an eye-catching manner. Let's start by placing text on the plot:

Use the `annotate()` method:

`annotate('Text Annotation String', xy=(x_placement_value, y_placement_value))`

The `xy` parameter tells matplotlib where to put your text on your graph. The x and y placement values are based on the values on the x and y axes. In other words, to add an annotation under the rocket line on the graph, you'd have the xy=(78, 720).


For example, in lieu of a legend, we can demonstrate placing an annotation alongside the line it describes:  

```python
# line plot for Rocket
plt.plot(stocks['day'],stocks['rocket'],color='b',label='Rocket')

# add the rocket label annotation
plt.annotate('Rocket', xy=(78, 720))

# line plot for Biotri
plt.plot(stocks['day'],stocks['biotri'],color='red',label='Biotri')

# add the biotri label annotation
plt.annotate('Biotri', xy=(78, 240))

# line plot for Wheatpharm
plt.plot(stocks['day'],stocks['wheatpharm'],color='#00FF00',label='Wheatpharm')

# add the wheatpharm label annotation
plt.annotate('Wheatpharm', xy=(90, 50))

# plot title
plt.title("Comparison of Wheatpharm, Rocket & Biotri")

# label x-axis
plt.xlabel("Days (since IPO)")

# label y-axis
plt.ylabel("Price (USD)")

# display figure
plt.show()
```

You can also add arrows with the `arrowprops` parameter. Adding `arrowprops` changes the placement parameters somewhat, as you can see here:

`annotate('Text Annotation String', xy=(placement_x_value, placement_y_value), xytext=(placement_x_value, placement_y_value), arrowprops=dict(property_name1= property_value1, property_name2= property_value2)`


1. The `xy` parameter specifies the point that the arrow points to. 

2. A new parameter called `xytext` specifies the location of the text.

3. The `arrowprops` parameter takes a dictionary whose keys are the property names and whose values are the property values. Note this dictionary doesn't look like regular python dictionaries (e.g., no {}). The parameter was built to turn this information into dictionary format for you. You just need to give it the right information. There are lots of possibilities, here's some basic ones:


| arrowprops | property names  | property values |
| - | -------------- | --------------- |
|   | width          | float    |
|   | headwidth      | float    |
|   | facecolor      | "Color"    |
|   | edgecolor      | "Color"    |
|   | linestyle      | {'-', '--', '-.', ':'} |


In the code below, we'll use arrows and text annotations on our "Comparison of Wheatpharm, Rocket & Biotri" graph to highlight:
1. When Rocket announced their new CEO, their stock price dropped while the other two companies' stock prices rose.
2. Biotri's all-time high stock price occurred as part of that bump.

```python
# Rocket line plot
plt.plot(stocks['day'],stocks['rocket'],color='b',label='Rocket')

# Biotri line plot
plt.plot(stocks['day'],stocks['biotri'],color='red',label='Biotri')

# Wheatpharm line plot 
plt.plot(stocks['day'],stocks['wheatpharm'],color='#00FF00',label='Wheatpharm')

# add the Biotri All-time High annotations
plt.annotate('Biotri All-time High', xy=(169, 520), xytext=(142, 700), arrowprops=dict(facecolor='orange', width=7))

# add the Rocket New CEO Annoucement annotations
plt.annotate('Rocket New CEO Annoucement', xy=(115, 740), xytext=(20, 600), arrowprops=dict(facecolor='r', width=3))

# title plot
plt.title("Comparison of Wheatpharm, Rocket & Biotri")

# place legend to center left
plt.legend(loc='center left')

# label x-axis
plt.xlabel("Days (since IPO)")

# label y-axis
plt.ylabel("Price (USD)")

# display figure
plt.show()
```
